CHECKING VIF

```python
from scipy import stats
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
def cal_vif(x):
    vif=pd.DataFrame()
    vif['Variables']=x.columns
    vif['VIF_FACTOR']=[variance_inflation_factor(x.values,i)for i in range(x.shape[1])]
    return(vif)
```

```python
cal_vif(x)
```

LINEAR REGRESSION 

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error,mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
```

```python
lr=LinearRegression()
```

```python
for i in range(0,100):
    x_train,x_test,y_train,y_test=train_test_split(x1,y2,test_size=0.20,random_state=i)
    lr.fit(x_train,y_train)
    predr=lr.predict(x_train)
    pred=lr.predict(x_test)
    print('At random_state' ,{i}, 'the training accuracy test is', r2_score(y_train,predr))
    print('At random_state' ,{i}, 'the testing accuracy test is', r2_score(y_test,pred))
    print('\n')
```

```python
x2=scale.fit_transform(x2)
```

```python
for i in range(0,100):
    x_train,x_test,y_train,y_test=train_test_split(x2,y2,test_size=0.20,random_state=i)
    lr.fit(x_train,y_train)
    predr=lr.predict(x_train)
    pred=lr.predict(x_test)
    print('At random_state' ,{i}, 'the training accuracy test is', r2_score(y_train,predr))
    print('At random_state' ,{i}, 'the testing accuracy test is', r2_score(y_test,pred))
    print('\n')
```
